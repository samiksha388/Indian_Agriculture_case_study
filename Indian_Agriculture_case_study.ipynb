```python
import pandas as pd

# Load the CSV file
file_path = 'IAS.csv'  # Replace with the actual file path
df = pd.read_csv(file_path)

# Find the crop with the maximum production
max_production_crop = df.loc[df['Production'].idxmax()]

# Find the crop with the minimum production
min_production_crop = df.loc[df['Production'].idxmin()]

print("Crop with maximum production:")
print(max_production_crop)

print("\nCrop with minimum production:")
print(min_production_crop)
```

```python
# Group by State_Name and sum the Production
state_production = df.groupby('State_Name')['Production'].sum()

# Find the state with maximum production
max_production_state = state_production.idxmax()
max_production_value = state_production.max()

# Find the state with minimum production
min_production_state = state_production.idxmin()
min_production_value = state_production.min()

print(f"State with maximum production: {max_production_state} with production {max_production_value}")
print(f"State with minimum production: {min_production_state} with production {min_production_value}")
```

```python
# Create a DataFrame for min and max production states
production_data = pd.DataFrame({
    'State': [min_production_state, max_production_state],
    'Production': [min_production_value, max_production_value]
})

# Plot the bar graph
production_data.plot(kind='bar', x='State', y='Production', legend=False)
plt.xlabel('State')
plt.ylabel('Production')
plt.title('Min and Max Production of the State')
plt.show()
```
